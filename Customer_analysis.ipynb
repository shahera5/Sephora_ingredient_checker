```python
import pandas as pd
```

```python
file_path = 'product_info.csv'
```

```python
filtered_chunks = []

# Read the CSV file in chunks of 10000 rows
chunk_size = 10000
term = 'Parfum'
for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    # Filter rows containing the specific term in a particular column
    filtered_chunk = chunk[chunk['ingredients'].str.contains(term, case=False, na=False)]
    filtered_chunks.append(filtered_chunk)

# Concatenate all filtered chunks into a single DataFrame
filtered_data = pd.concat(filtered_chunks, ignore_index=True)

print(filtered_data)
```
